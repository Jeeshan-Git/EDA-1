## EDA

# 1. Import necessary libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
df = pd.read_csv("C:\\Data science\\Project\\Ecommerce_Customers (2).csv")
```

# 2. Dataset Overview

```python
# Displays structure of the dataset
print(df.info())
```

```python
# Summary statistics for numerical columns
print(df.describe())
```

# 3. Check for missing values

```python
missing_values = df.isnull().sum()  # Check missing values per column
print(missing_values)
```

# 4. Correlation matrix

```python
numerical_columns = df.select_dtypes(include=[np.number])  # Select only numeric columns
correlation_matrix = numerical_columns.corr()  # Calculate correlation matrix
print(correlation_matrix)
```

```python
# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

# 5. Visualizations

```python


# 5.1 Distribution plots (histograms) for individual features
plt.figure(figsize=(18, 12))

plt.subplot(3, 2, 1)
sns.histplot(df['Avg Session Length'], kde=True, color='blue')
plt.title('Distribution of Avg Session Length')

plt.subplot(3, 2, 2)
sns.histplot(df['Time on App'], kde=True, color='green')
plt.title('Distribution of Time on App')

plt.subplot(3, 2, 3)
sns.histplot(df['Time on Website'], kde=True, color='red')
plt.title('Distribution of Time on Website')

plt.subplot(3, 2, 4)
sns.histplot(df['Yearly Amount Spent'], kde=True, color='purple')
plt.title('Distribution of Length of Membership')

plt.subplot(3, 2, 5)
sns.histplot(df['Yearly Amount Spent'], kde=True, color='orange')
plt.title('Distribution of Yearly Amount Spent')

plt.tight_layout()
plt.show()
```

# 5.2 Boxplots to check for outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Avg Session Length', 'Time on App', 'Time on Website', 'Yearly Amount Spent']], palette="Set2")
plt.title("Boxplot of Numerical Columns")
plt.show()
```

# 5.3 Pairplot to visualize relationships between features

```python
sns.pairplot(df[['Avg Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent']])
plt.title("Pairplot of Features")
plt.show()
```

# 6. Check for correlations between target (Yearly Amount Spent) and other features

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Avg Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation with Yearly Amount Spent")
plt.show()
```
